```python
# Header import libraries
# Give path to the downloadd Genbank file

from Bio import SeqIO
path = "/content/genbanksequence.gb"
genbank = SeqIO.read(path, "genbank")
```

```python
# For more data extraction for the Genbank file

id = genbank.id
print(id)
```

```python
description = genbank.description
print(description)
```

```python
sequences = genbank.seq
seq_length = len(sequences)
print(seq_length)
```

```python
# Feature extraction from a Genbank file

features = genbank.features
feature_types = [ feature.type for feature in features ]
feature_types = set(feature_types)
feature_types = list(feature_types)
print(feature_types)
```

```python
# Iterating features from all the features

feature_types = ['sig_peptide', 'precursor_RNA', 'J_segment', 'mRNA', 'protein_bind', 'misc_recomb', 'misc_RNA', 'tRNA', 'ncRNA', 'regulatory', 'exon', 'transit_peptide', 'rep_origin', 'source', 'V_segment', 'centromere', 'misc_feature', 'CDS', 'repeat_region', 'mobile_element', 'C_region', 'gene', 'mat_peptide', 'proprotein']
print(feature_types)
```

```python
# Use of for loop for iterate number of feature count in the feature list.
# Use of pipe operator

for feature in feature_types:
  allfeatures=[i for i in features if i.type==feature ]
  No_of_features=len(allfeatures)
  print('%s:%d'%(feature, No_of_features) )
```
